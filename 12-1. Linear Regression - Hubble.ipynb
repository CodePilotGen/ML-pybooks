```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

### Задача простой линейной регрессии
$$
y \approx \beta_0 + \beta_1 x,
$$
т.е.
$$
y^{(i)} \approx \beta_0 + \beta_1 x^{(i)}
\qquad
(i=1,2,\dots, N).
$$

А именно, ищем $\beta_0$, $\beta_1$, на которых достигается минимум
$$
\min_{\beta_0,\, \beta_1} (y^{(i)} - \beta_0 - \beta_1 x^{(i)})^2
$$

### Задача множественной линейной регрессии
$$
y \approx \beta_0 + \beta_1 x_1 + \beta_2 x_2 + \dots +\beta_d x_d,
$$
т.е.
$$
y^{(i)} \approx \beta_0 + \beta_1 x_1^{(i)} + \beta_2 x_2^{(i)} + \dots +\beta_d x_d^{(i)}
\qquad
(i=1,2,\dots, N).
$$
А именно, ищем $\beta_0, \beta_1, \dots, \beta_d$, на которых достигается минимум
$$
\min_{\beta_0,\dots, \beta_d} \left(y^{(i)} - \beta_0 - \sum_{j=1}^d \beta_j x_j^{(i)} \right)^2
$$

_Остаточная сумма квадратов_:
$$
{\rm RSS} = \sum_{i=1}^N \left(y^{(i)} - \beta_0 - \sum_{j=1}^d \beta_j x_j^{(i)} \right)^2 \to \min
$$
_Средняя квадратическая ошибка_:
$$
{\rm MSE} = \frac{1}{N} {\rm RSS} \to \min
$$

```python
data = np.loadtxt('../source/hubble.dat', delimiter = "\t", dtype = float, skiprows = 1)
x = data[:, [0]]
y = data[:, 1]
```

```python
plt.scatter(x, y, color = 'b')
plt.xlabel('Distance from Earth (Megaparsecs)')
plt.ylabel('The recesson velocity (km/sec)')
plt.title('E. Hubble\'s data')
```

```python
from sklearn.linear_model import LinearRegression
lr_model = LinearRegression()
lr_model.fit(x, y)
beta_0 = lr_model.intercept_
beta_1 = lr_model.coef_[0]
print(beta_0, beta_1)
```

```python
plt.scatter(x, y, color = 'b')
plt.xlabel('Distance from Earth (Megaparsecs)')
plt.ylabel('The recesson velocity (km/sec)')
plt.title('E. Hubble\'s data')
import mysk
mysk.abline(beta_0, beta_1)
```

```python
lr_model = LinearRegression(fit_intercept = False)
lr_model.fit(x, y)
beta_0 = lr_model.intercept_
beta_1 = lr_model.coef_[0]
print(beta_0, beta_1)
```
